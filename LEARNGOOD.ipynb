```python
f = open("НАШИФАЙЛЫ/BADCLEAR.txt", 'r')
melk = []
for line in f:
    for i in line.strip("\n").split(' '):
        if i not in melk: 
            melk.append(i)
f.close()
```

```python
print(melk)
```

```python
f = open("НАШИФАЙЛЫ/GOODCLEAR.txt", 'r')
melkg = []
for line in f:
    for i in line.strip("\n").split(' '):
        if i not in melkg: 
            melkg.append(i)
f.close()
```

```python
print(melkg)
```

```python
melkotl = []
melksovp = []
schsotl = []
schsovp = []
for i in melk:
    elemot = re.sub(r"[1234567890]",' ', str(i))
    elemot = ' '.join(elemot.split())
    if (elemot not in melkg) and (elemot != '') and (elemot not in melkotl):
        melkotl.append(elemot)
        schsotl.append(1)
    if (elemot not in melkg) and (elemot in melkotl) and (elemot != ''):
        schsotl[melkotl.index(elemot)] += 1
    if (elemot in melkg) and (elemot != '') and (elemot not in melksovp):
        melksovp.append(elemot)
        schsovp.append(1)
    if (elemot in melkg) and (elemot in melksovp) and (elemot != ''):
        schsovp[melksovp.index(elemot)] += 1
```
